```python
# -------------------------------
# Task 1: User Login Check
# -------------------------------

# Predefined credentials
correct_username = "admin"
correct_password = "1234"

# Input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Checking credentials
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# -------------------------------
# Task 2: Pass / Fail Analyzer
# -------------------------------

# Given marks list
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Loop through marks and count pass/fail
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Students:", len(marks))
print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

```python
# -------------------------------
# Task 3: Simple Data Cleaner
# -------------------------------

# Given names list
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Cleaning process
for name in names:
    cleaned_name = name.strip().lower()  # Remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

# Display cleaned list
print("Cleaned Names List:", cleaned_names)
```

```python
# -------------------------------
# Task 4: Message Length Analyzer
# -------------------------------

# Given messages list
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    # Flag if message length > 10
    if length > 10:
        print("⚠️ This message is longer than 10 characters.")
```
