```python
# ============================================
# ASSIGNMENT 5 – Real-World Function Problems
# Data Science Internship – February 2026
# ============================================

# --------------------------------------------
# Problem 1: Social Media – Post Engagement Analyzer
# --------------------------------------------

def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like
    
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example
analyze_post_engagement([200, 300, 250, 300])


# --------------------------------------------
# Problem 2: Healthcare – Medicine Stock Alert System
# --------------------------------------------

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example
check_medicine_stock(6)


# --------------------------------------------
# Problem 3: Agriculture – Rainfall Adequacy Checker
# --------------------------------------------

def check_rainfall(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example
check_rainfall([70, 75, 68, 80, 67], 70)


# --------------------------------------------
# Problem 4: Social Media – Duplicate Account Detection
# --------------------------------------------

def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)


# Example
detect_duplicate_accounts(["user1", "user2", "user1"])


# --------------------------------------------
# Problem 5: Healthcare – Appointment Eligibility Checker
# --------------------------------------------

def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Example
check_appointment_eligibility(21)


# --------------------------------------------
# Problem 6: Agriculture – Premium Crop Price Filter
# --------------------------------------------

def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example
filter_premium_crops([1500, 2500, 1800, 3200])


# --------------------------------------------
# Problem 7: System Monitoring – Application Health Checker
# --------------------------------------------

def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)


# Example
check_system_health(7)


# --------------------------------------------
# Problem 8: Banking – Daily Transaction Limit Checker
# --------------------------------------------

def check_transaction(amount):
    limit = 50000
    
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example
check_transaction(60000)


# --------------------------------------------
# Problem 9: E-Learning – Student Attendance Eligibility
# --------------------------------------------

def check_attendance(attended_classes, total_classes):
    percentage = (attended_classes / total_classes) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example
check_attendance(40, 50)


# --------------------------------------------
# Problem 10: Smart Electricity Bill Analyzer
# --------------------------------------------

def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage


# Example
bill_amount, usage_status = calculate_electricity_bill(250)
print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
