```python
def manage_parking(capacity, vehicle_logs):
    parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak_usage:
            peak_usage = parked

    if parked > capacity:
        status = "Over Capacity Alert"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak_usage)
    print("Parking Status:", status)


# Example
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km

    # Traffic adjustment
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10
    else:
        base_time += 5

    # Weather adjustment
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Example
estimate_delivery_time(8, "High", "Rainy")
```

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", str(int(occupancy)) + "%")
    print("Show Status:", status)


# Example
analyze_occupancy(200, [1] * 150)
```

```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", str(int(average)) + "%")
    print("Server Status:", status)


# Example
classify_server_load([45, 60, 70, 85, 90])
```

```python
def monitor_resources(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)


# Example
monitor_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def manage_registrations(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Example
manage_registrations(100, 105)
```
