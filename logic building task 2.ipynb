```python
# -------------------------------------------------
# Problem 1: Unique Words in a Sentence
# -------------------------------------------------

# Taking sentence input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# -------------------------------------------------
# Problem 2: Highest Salary from Employee Data
# -------------------------------------------------

# Employee salary dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Display result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
# -------------------------------------------------
# Problem 3: Find Maximum and Minimum Values
# -------------------------------------------------

numbers = [45, 22, 89, 10, 66]

# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

# Loop through numbers
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# -------------------------------------------------
# Problem 4: Count Products Above a Price Threshold
# -------------------------------------------------

prices = [450, 1200, 899, 1500, 300]

count = 0

# Checking prices above 1000
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

```python
# -------------------------------------------------
# Problem 5: Calculate Attendance Percentage
# -------------------------------------------------

attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = attendance.count("P")

# Calculate percentage
attendance_percentage = (present_count / len(attendance)) * 100

# Display result
print("Attendance Percentage:", attendance_percentage)
```

```python
# -------------------------------------------------
# Problem 6: Remove Duplicate Phone Numbers
# -------------------------------------------------

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set
unique_numbers = set(phone_numbers)

# Display result
print("Unique phone numbers:", unique_numbers)
```

```python
# -------------------------------------------------
# Problem 7: Count Character Frequency
# -------------------------------------------------

text = "python"

char_frequency = {}

# Counting each character
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

# Display result
print(char_frequency)
```

```python
# -------------------------------------------------
# Problem 8: Convert List to Tuple
# -------------------------------------------------

sample_list = [10, 20, 30]

# Convert to tuple
sample_tuple = tuple(sample_list)

# Display result
print("Tuple:", sample_tuple)
```

```python
# -------------------------------------------------
# Problem 9: Check if a Key Exists in a Dictionary
# -------------------------------------------------

employees = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

# Check key existence
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# -------------------------------------------------
# Problem 10: Calculate Average Marks
# -------------------------------------------------

marks = [70, 80, 65, 75]

# Calculate average
average_marks = sum(marks) / len(marks)

# Display result
print("Average Marks:", average_marks)
```
